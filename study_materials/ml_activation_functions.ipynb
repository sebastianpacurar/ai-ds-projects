# NB Setup

```python
import numpy as np
import matplotlib.pyplot as plt


def plot_func(x, y, equation, title):
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label=equation, lw=3)
    plt.xlabel("$x$")
    plt.ylabel("$f(x)$")
    plt.figtext(0.5, 1, title, fontsize=20, ha='center')
    plt.figtext(0.5, 0.9, equation, fontsize=25, ha='center')
    plt.subplots_adjust(top=0.75)
    plt.grid(True, lw=0.5, color="0.8")
    plt.show()
```

## 1.1 - Sigmoid
* Converts outputs into probabilities, in logistic regression
* Used as an activation function in early models, in neural netwirks
* Probability Estimation: Maps any real number to (0,1)

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))


x = np.linspace(-10, 10, 100)
y = sigmoid(x)

plot_func(x, y, equation=r"$f(x) = \frac{1}{1 + e^{-x}}$", title="Sigmoid Function")
```

<br><hr><br>

## 1.2 - Tanh
* Neural Networks → Used as an activation function in newural networks
* Better than Sigmoid, because outputs are in the range (-1,1), meaning that it's zero-centered

```python
def tanh(x):
    return np.tanh(x)


x = np.linspace(-10, 10, 100)
y = tanh(x)

plot_func(x, y, equation=r"$f(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$", title="Tanh Function")
```

<br><hr><br>

## 1.3 - Hard Sigmoid
* Best For: Situations where computational efficiency is critical
* Key Advantage: Faster computation compared to the standard Sigmoid function
* Limitation: Less smooth and more prone to approximation errors

```python
def hard_sigmoid(x):
    return np.clip(0.2 * x + 0.5, 0, 1)


x = np.linspace(-10, 10, 100)
y = hard_sigmoid(x)

plot_func(x, y, equation=r"$f(x) = \max(0, \min(1, 0.2x + 0.5))$", title="Hard Sigmoid Function")
```

<br><hr><br>

## 1.4 - Log Sigmoid

* Best For: Certain ML models, probabilistic tasks
* Key Advantage: Logarithmic scaling of the Sigmoid function
* Limitation: Computationally inefficient

```python
def log_sigmoid(x):
    return -np.log1p(np.exp(-x))  # equivalent to log(1 / (1 + exp(-x)))


x = np.linspace(-10, 10, 100)
y = log_sigmoid(x)

plot_func(x, y, equation=r"$f(x) = \ln \left( \frac{1}{1 + e^{-x}} \right)$", title="Log Sigmoid Function")
```

<br><hr><br>

## 1.5 - Softmax
* Used in the output layer of classification models (ex: deep learning)
* Converts a vector into a probability distribution
* Helps multiclass classification tasks

```python
def softmax(x):
    exp_x = np.exp(x - np.max(x))
    return exp_x / exp_x.sum()


x = np.linspace(-4, 4, 100)
y = softmax(x)

plot_func(x, y, equation=r"$f(x)_i = \frac{e^{x_i}}{\sum_{j=1}^{n} e^{x_j}}$", title="Softmax Function")
```

## 2.1 - ReLU (Rectified Linear Unit)
* Most widely used activation function in deep learning
* Solves vanishing gradient issue (for positive values)
* Efficient & Simple leading to faster computation

```python
def relu(x):
    return np.maximum(0, x)


x = np.linspace(-10, 10, 100)
y = relu(x)

plot_func(x, y, equation=r"$f(x) = \max(0, x)$", title="ReLU Activation Function")
```

<br><hr><br>

## 2.2 - Leaky ReLU
* Fixes "Dying ReLU" problem, allowing small negative gradients
* Used in deep learning architectures for stable training

```python
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)


x = np.linspace(-10, 10, 100)
y = leaky_relu(x)

plot_func(x, y, equation=r"$f(x) = \max(0.01x, x)$", title="Leaky ReLU Activation Function")
```

<br><hr><br>

## 2.3 - RReLU (Randomized ReLU)

* Best For: Avoiding dying ReLU in CNNs
* Key Advantage: Introduces randomization to prevent neuron dying
* Limitation: Slightly increases training variance

```python
def rrelu(x, alpha=0.1, beta=0.3):
    return np.where(x > 0, x, np.random.uniform(alpha, beta) * x)


x = np.linspace(-10, 10, 100)
y = rrelu(x)

plot_func(x, y, equation=r"$f(x) = x \text{ if } x \geq 0, \quad \alpha x \text{ if } x < 0, \quad \alpha \sim U(l, u)$", title="RReLU Function")
```

<br><hr><br>

## 2.4 - PReLU (Parametric ReLU)

* Best For: CNNs & Deep Learning
* Key Advantage: Learns the negative slope parameter during training
* Limitation: Extra parameters increase model complexity

```python
def prelu(x, alpha=0.1):
    return np.where(x > 0, x, alpha * x)


x = np.linspace(-10, 10, 100)
y = prelu(x)

plot_func(x, y, equation=r"$f(x) = x \text{ if } x \geq 0, \quad \alpha x \text{ if } x < 0$", title="PReLU Function")
```

<br><hr><br>

## 2.5 - ELU
* Best For: Deep learning
* Key Advantage: Helps avoid dead neurons; output can be negative, making it zero-centered
* Limitation: Computationally expensive for large datasets

```python
def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))


x = np.linspace(-5, 5, 1000)
y = elu(x)

plot_func(x, y, equation=r"$f(x) = \max(0, x) \text{ or } \alpha (e^x - 1)$", title="Exponential Linear Unit (ELU)")
```

<br><hr><br>

## 2.6 - SELU
* Best For: Self-normalizing networks
* Key Advantage: Self-normalizing, improving training stability
* Limitation: Sensitive to initialization and scaling of input data

```python
# λ and α are specific constants
def selu(x, alpha=1.6733, lambda_=1.0507):
    return lambda_ * np.where(x > 0, x, alpha * (np.exp(x) - 1))


x = np.linspace(-10, 10, 100)
y = selu(x)

plot_func(x, y, equation=r"$f(x) = \lambda \max(0, x) \text{ or } \alpha (e^x - 1)$", title="Scaled Exponential Linear Unit (SELU)")
```

<br><hr><br>

## 2.7 - Softplus

* Best For: Replacement for ReLU (less sparse)
* Key Advantage: Smooth, differentiable function
* Limitation: Slower than ReLU and computationally expensive

```python
def softplus(x):
    return np.log(1 + np.exp(x))


x = np.linspace(-10, 10, 100)
y = softplus(x)

plot_func(x, y, equation=r"$f(x) = \log(1 + e^x)$", title="Softplus Activation Function")
```

## 3.1 - Maxout


* Best For: Generalized ReLU
* Key Advantage: Can approximate any convex function
* Limitation: Requires additional parameters

```python
def maxout(x):
    # simplified, since maxout usually involves multiple learned weights
    return np.maximum(x, 0)


x = np.linspace(-10, 10, 100)
y = maxout(x)

plot_func(x, y, equation=r"$f(x) = \max(0, x)$", title="Maxout Function")
```

<br><hr><br>

## 3.2 - Swish

```python
def swish(x):
    return x * sigmoid(x)


x = np.linspace(-10, 10, 100)
y = swish(x)

plot_func(x, y, equation=r"$f(x) = x \cdot \sigma(x)$", title="Swish Activation Function")
```

<br><hr><br>

## 3.3 - Swish-1

* Best For: Deep Learning
* Key Advantage: Smooth, better than ReLU in some cases
* Limitation: Computationally slower

```python
def swish_1(x):
    return x / (1 + np.exp(-x))


x = np.linspace(-10, 10, 100)
y = swish_1(x)

plot_func(x, y, equation=r"$f(x) = x \frac{1}{1 + e^{-x}}$", title="Swish-1 Function")
```

<br><hr><br>

## 3.4 - Mish (Self-Gated Activation)

* Best For: Deep Learning, CNNs, NLP models
* Key Advantage: Similar to Swish but can outperform ReLU
* Limitation: Computationally expensive

```python
def mish(x):
    return x * np.tanh(np.log1p(np.exp(x)))  # log1p(x) = log(1+x) for numerical stability


x = np.linspace(-10, 10, 100)
y = mish(x)

plot_func(x, y, equation=r"$f(x) = x (\tanh(\ln(1 + e^x)))$", title="Mish Function")
```

# 4) Specialized & Less Common Functions

<br><hr><br>

## 4.1 - GELU
* Best For: Transformer models (ex: BERT, GPT)
* Key Advantage: Smooth and differentiable; less prone to vanishing gradient
* Limitation: More complex than ReLU, leading to slower computations

```python
def gelu(x):
    return 0.5 * x * (1 + np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * x ** 3)))


x = np.linspace(-10, 10, 100)
y = gelu(x)

plot_func(x, y, equation=r"$f(x) = 0.5x (1 + \tanh(\sqrt{\frac{2}{\pi}} (x + 0.044715 x^3)))$", title="Gaussian Error Linear Unit (GELU)")
```

<br><hr><br>

## 4.2 - Bent Identity

* Best For: Hybrid activation approaches
* Key Advantage: Similar to ReLU but smooth and differentiable everywhere
* Limitation: Less commonly used, lacks extensive research

```python
def bent_identity(x):
    return (np.sqrt(x ** 2 + 1) - 1) / 2 + x


x = np.linspace(-10, 10, 100)
y = bent_identity(x)

plot_func(x, y, equation=r"$f(x) = \frac{\sqrt{x^2 + 1} - 1}{2} + x$", title="Bent Identity Function")
```

<br><hr><br>

## 4.3 - Hard Tanh

* Best For: Efficiency-critical tasks
* Key Advantage: Faster approximation of Tanh
* Limitation: Loses smoothness, introduces approximation error

```python
def hard_tanh(x):
    return np.clip(x, -1, 1)


x = np.linspace(-10, 10, 100)
y = hard_tanh(x)

plot_func(x, y, equation=r"$f(x) = \max(-1, \min(1, x))$", title="Hard Tanh Function")
```

<br><hr><br>

## 4.4 - Sinusoidal (Sine Activation)

* Best For: Function approximation & physics-based neural networks
* Key Advantage: Periodic, useful in some specialized tasks
* Limitation: Not widely used in standard deep learning

```python
def sinusoidal(x):
    return np.sin(x)


x = np.linspace(-10, 10, 100)
y = sinusoidal(x)

plot_func(x, y, equation=r"$f(x) = \sin(x)$", title="Sinusoidal Function")
```
