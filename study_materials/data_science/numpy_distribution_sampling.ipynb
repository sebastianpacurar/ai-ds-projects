```python
import numpy as np
import matplotlib.pyplot as plt


def plot_hist(graph_title, datasets, titles, rows, cols, bins):
    fig, axes = plt.subplots(rows, cols, figsize=(16, 10))
    fig.suptitle(graph_title, fontsize=16, fontweight="bold")

    for ax, data, title in zip(axes.flat, datasets, titles):
        ax.hist(data, bins=bins, alpha=0.6, color="b", edgecolor="black")
        ax.set_xlabel("Value")
        ax.set_ylabel("Frequency")
        ax.text(0.5, 1, title, transform=ax.transAxes,
                ha="center", va="bottom", fontsize=12, fontweight="bold")
        ax.grid(axis="y", lw=0.5, ls="-.", color="0.5")

    plt.tight_layout()
    plt.show()
```

# uniform()

Generates a ***uniform ditribution***

Takes 3 arguments:
* `low` = min number. defaults to ***0.0***
* `high` = max number. defaults to ***1.0***
* `size` = the size of the array. defaults to None

Use cases:
* ***Random float generation with a fixed range***

```python
ud_1 = np.random.uniform(low=0, high=10, size=100)  # standard distribution
ud_2 = np.random.uniform(low=-10, high=10, size=100)  # symmetric distribution around 0
ud_3 = np.random.uniform(low=5, high=15, size=100)  # shifted to the right
ud_4 = np.random.uniform(low=-5, high=5, size=100)  # shifted to the left
ud_5 = np.random.uniform(low=0, high=20, size=100)  # wider range
ud_6 = np.random.uniform(low=0, high=5, size=100)  # narrower range
ud_7 = np.random.uniform(low=-15, high=5, size=100)  # negative low and positive high
ud_8 = np.random.uniform(low=-20, high=-10, size=100)  # negative range

datasets = [ud_1, ud_2, ud_3, ud_4, ud_5, ud_6, ud_7, ud_8]
titles = [
    "low=0, high=10",
    "low=-10, high=10",
    "low=5, high=15",
    "low=-5, high=5",
    "low=0, high=20",
    "low=0, high=5",
    "low=-15, high=5",
    "low=-20, high=-10"
]

plot_hist(graph_title="Uniform Distribution",
          rows=2, cols=4, bins=20, datasets=datasets, titles=titles)
```

# binomial()

* Models the number of successes in `n` trials with success probability `p`, where each trial has:
    * ***Probability of success*** = `p`
    * ***Probability of failure*** = `1 - p`

*  Use cases:
    * ***Coin flips***
    * ***Quality control***
    * ***Decision-making models***

```python
bd_1 = np.random.binomial(n=5, p=0.5, size=100)  # standard distribution, moderate n
bd_2 = np.random.binomial(n=10, p=0.5, size=100)  # standard distribution
bd_3 = np.random.binomial(n=20, p=0.5, size=100)  # higher number of trials (n)
bd_4 = np.random.binomial(n=10, p=0.3, size=100)  # lower probability of success
bd_5 = np.random.binomial(n=10, p=0.7, size=100)  # higher probability of success
bd_6 = np.random.binomial(n=15, p=0.4, size=100)  # moderate n, lower p
bd_7 = np.random.binomial(n=5, p=0.7, size=100)  # lower n, higher p
bd_8 = np.random.binomial(n=30, p=0.6, size=100)  # larger n with moderate p

datasets = [bd_1, bd_2, bd_3, bd_4, bd_5, bd_6, bd_7, bd_8]
titles = [
    "n=5, p=0.5",
    "n=10, p=0.5",
    "n=20, p=0.5",
    "n=10, p=0.3 (Lower p)",
    "n=10, p=0.7 (Higher p)",
    "n=15, p=0.4",
    "n=5, p=0.7 (Higher p)",
    "n=30, p=0.6 (Larger n)"
]

plot_hist(graph_title="Binomial Distribution",
          rows=2, cols=4, bins=20, datasets=datasets, titles=titles)
```

    * `p` = the probability of success on each trial

* Use cases:
    * ***Modeling the trials needed for success***, such as:
        * ***Failure tests***
        * ***Performance evaluations***


```python
nbd_1 = np.random.negative_binomial(n=5, p=0.5, size=100)  # standard distribution, moderate n
nbd_2 = np.random.negative_binomial(n=10, p=0.5, size=100)  # standard distribution
nbd_3 = np.random.negative_binomial(n=20, p=0.5, size=100)  # higher number of successes
nbd_4 = np.random.negative_binomial(n=10, p=0.3, size=100)  # lower probability of success
nbd_5 = np.random.negative_binomial(n=10, p=0.7, size=100)  # higher probability of success
nbd_6 = np.random.negative_binomial(n=15, p=0.4, size=100)  # moderate n, lower p
nbd_7 = np.random.negative_binomial(n=5, p=0.7, size=100)  # lower n, higher p
nbd_8 = np.random.negative_binomial(n=30, p=0.6, size=100)  # large n with moderate p

datasets = [nbd_1, nbd_2, nbd_3, nbd_4, nbd_5, nbd_6, nbd_7, nbd_8]
titles = [
    "n=5, p=0.5",
    "n=10, p=0.5",
    "n=20, p=0.5",
    "n=10, p=0.3 (Lower p)",
    "n=10, p=0.7 (Higher p)",
    "n=15, p=0.4",
    "n=5, p=0.7 (Higher p)",
    "n=30, p=0.6 (Larger n)"
]

plot_hist(graph_title="Negative Binomial Distribution",
          rows=2, cols=4, bins=20, datasets=datasets, titles=titles)
```

# multinomial()

* Generates random values from the multinomial distribution, which ***generalizes the binomial distribution*** for more than two outcomes
* It models the counts of various categories in a series of trials
* Takes 3 arguments:
    * `n` = the number of trials
    * `pvals` = an array of probabilities for each outcome
    * `size` = the number of random values to generate
* Use cases:
    * ***Modeling categorical outcomes***, such as:
        * ***Survey responses***
        * ***Gene sequencing***

```python
multinomial_data = np.random.multinomial(n=10, pvals=[0.2, 0.3, 0.5], size=10000)
plt.figure(figsize=(8, 4))
plt.hist(multinomial_data, label=['Category 1', 'Category 2', 'Category 3'], stacked=True)
plt.title("Multinomial Distribution")
plt.xlabel("Category")
plt.ylabel("Frequency")
plt.legend()
plt.grid()
plt.show()
```

# poisson()
* Models the number of events in a fixed time interval
* Takes 2 arguments:
    * `lam` (Î») = expected number of events per interval
    * `size` = number of samples
* Use cases:
    * ***Call center calls per hour***
    * ***Event modeling***:
        * ***Customer arrivals***
        * ***Traffic accidents***

```python
poisson_1 = np.random.poisson(lam=1, size=100)  # lower rate (more variance)
poisson_2 = np.random.poisson(lam=5, size=100)  # standard rate
poisson_3 = np.random.poisson(lam=10, size=100)  # higher rate (more concentration around mean)
poisson_4 = np.random.poisson(lam=15, size=100)  # even higher rate (less spread)
poisson_5 = np.random.poisson(lam=20, size=100)  # more peaked around the mean
poisson_6 = np.random.poisson(lam=0.5, size=100)  # lower rate (more spread out)
poisson_7 = np.random.poisson(lam=25, size=100)  # very high rate
poisson_8 = np.random.poisson(lam=50, size=100)  # extremely high rate

datasets = [poisson_1, poisson_2, poisson_3, poisson_4, poisson_5, poisson_6, poisson_7, poisson_8]
titles = [
    "Î»=1 (Low Rate)",
    "Î»=5 (Standard Rate)",
    "Î»=10 (Higher Rate)",
    "Î»=15 (Even Higher Rate)",
    "Î»=20 (More Peaked)",
    "Î»=0.5 (Very Low Rate)",
    "Î»=25 (Very High Rate)",
    "Î»=50 (Extremely High Rate)"
]

plot_hist(graph_title="Poisson Distribution",
          rows=2, cols=4, bins=20, datasets=datasets, titles=titles)
```

# standard_t()

* Generates random values from the ***Student's t-distribution***, used in hypothesis testing and when sample sizes are small
* Takes 2 arguments:
    * `df` = degrees of freedom, parameter that controls the distribution's shape
    * `size` = the number of random values to generate

```python
std_1 = np.random.standard_t(df=1, size=1000)  # really heavy tails
std_2 = np.random.standard_t(df=2, size=1000)  # heavy tails
std_3 = np.random.standard_t(df=3, size=1000)  # light tails
std_4 = np.random.standard_t(df=5, size=1000)  # lighter tails
std_5 = np.random.standard_t(df=10, size=1000)  # approaching normal
std_6 = np.random.standard_t(df=30, size=1000)  # approaching normal

datasets = [std_1, std_2, std_3, std_4, std_5, std_6]
titles = [
    "df=1 (Really Heavy Tails)",
    "df=2 (Heavy Tails)",
    "df=3 (Ligth Tails)",
    "df=5 (Lighter Tails)",
    "df=10 (Close to Normal)",
    "df=30 (Closer to Normal)",
]

plot_hist(graph_title="Student's t-Distribution",
          rows=2, cols=3, bins=50, datasets=datasets, titles=titles)
```

# standard_normal()
* Generates random numbers from a ***standard normal distribution***, which has:
    * Mean = 0
    * Standard Deviation = 1
* This is a special case of the normal distribution with fixed parameters (mean = 0, standard deviation = 1)
* This generates the same result as ***randn()***
* Takes 1 argument:
    * `size` = the number of random values to generate
* Use cases:
    * ***Simulations or modeling where you need data that follows a standard normal distribution***, such as:
        * ***Physics - Particle Movement (Brownian Motion)***
        * ***Biology - Genetic Variability***


```python
snd_1 = np.random.standard_normal(100)  # Standard normal (mean=0, std=1)
snd_2 = np.random.standard_normal(100) * 2  # Wider distribution (mean=0, std=2)
snd_3 = np.random.standard_normal(100) * 0.5  # Narrower distribution (mean=0, std=0.5)
snd_4 = np.random.standard_normal(100) + 1  # Shifted right (mean=1, std=1)
snd_5 = np.random.standard_normal(100) - 1  # Shifted left (mean=-1, std=1)
snd_6 = np.random.standard_normal(100) * 3  # Even wider distribution (mean=0, std=3)
snd_7 = np.random.standard_normal(100) + 2  # Shifted right (mean=2, std=1)
snd_8 = np.random.standard_normal(100) - 2  # Shifted left (mean=-2, std=1)

datasets = [snd_1, snd_2, snd_3, snd_4, snd_5, snd_6, snd_7, snd_8]
titles = [
    "Standard Normal (mean=0, std=1)",
    "Wider Distribution (std=2)",
    "Narrower Distribution (std=0.5)",
    "Shifted Right (mean=1)",
    "Shifted Left (mean=-1)",
    "Wider Distribution (std=3)",
    "Shifted Right (mean=2)",
    "Shifted Left (mean=-2)"
]

plot_hist(graph_title="Standard Normal Distribution",
          rows=2, cols=4, bins=15, datasets=datasets, titles=titles)
```

# beta()

* Models probabilities (values bwetween `0` and `1`)
* Takes 3 arguments:
    * `a` (alpha) = the shape parameter for the ***distribution's first part***. Controls the "left" side (the lower values)
    * `b` (beta) = the shape parameter for the ***distribution's second part***. Controls the "right" side (the higher values)
    * `size` = number of samples
* The shape of the distribution changes depending on the values of `a` and `b`:
    * When a = b = 1, the Beta distribution is a uniform distribution between 0 and 1.
    * When a > b, the distribution is ***skewed toward 1*** (more values near 1)
    * When a < b, the distribution is ***skewed toward 0*** (more values near 0)
    * When a = b > 1, the distribution has ***a bell-shaped curve*** centered around `0.5`
* Use cases:
    * ***Modeling probabilities***
    * ***Bayesian statistics***
    * ***A/B Testing***

```python
bd_1 = np.random.beta(a=0.5, b=0.5, size=1000)  # very skewed, long tail
bd_2 = np.random.beta(a=2, b=2, size=1000)  # symmetric distribution
bd_3 = np.random.beta(a=5, b=5, size=1000)  # sharper peak around 0.5
bd_4 = np.random.beta(a=1, b=3, size=1000)  # left-skewed
bd_5 = np.random.beta(a=3, b=1, size=1000)  # right-skewed
bd_6 = np.random.beta(a=10, b=2, size=1000)  # right-skewed with a sharp peak
bd_7 = np.random.beta(a=2, b=10, size=1000)  # left-skewed with a sharp peak
bd_8 = np.random.beta(a=10, b=10, size=1000)  # symmetric and sharp peak at 0.5

datasets = [bd_1, bd_2, bd_3, bd_4, bd_5, bd_6, bd_7, bd_8]
titles = [
    "a=0.5, b=0.5 (Skewed)",
    "a=2, b=2 (Symmetric)",
    "a=5, b=5 (Sharp Peak)",
    "a=1, b=3 (Left-Skewed)",
    "a=3, b=1 (Right-Skewed)",
    "a=10, b=2 (Right-Skewed with Sharp Peak)",
    "a=2, b=10 (Left-Skewed with Sharp Peak)",
    "a=10, b=10 (Symmetric)"
]

plot_hist(graph_title="Beta Distribution",
          rows=2, cols=4, bins=15, datasets=datasets, titles=titles)
```

# chisquare()

* Generates random values from the ***chi-squared distribution***
* Takes 2 arguments:
    * `df` = degrees of freedom; it's the shape parameter (sometimes denoted as `k`)
    * `size` = the number of values to generate
* Use cases:
    * ***Used in chi-squared tests***
